#### Creating Copy of original data Frame

```python
#copy() fucntion is use for copying the dataframe.
df1 = df.copy()
df1.head()
```

```python
#renaming the name column as brand
df1 = df1.rename(columns={'name':'brand'})
df1.head()
```

```python
#plot the graph of average sekking price of brand
ax   = a0.plot(kind='barh', cmap='PRGn' , figsize=(10,16) ,title= 'Avarege Selling Price of Car Brand')

#setting x-axis label
plt.xlabel("Average Selling Price",fontsize=10,color="black")

#setting y-axis label
plt.ylabel("Brand",fontsize=10,color="black")

for c in ax.containers:
        # set the bar label
    ax.bar_label(c, fmt='%.0f',label_type='center', color='w',rotation=0)
```

As per graph we can say that brands like **Land Rover, BMW, Mercedes-Benz, Volvo, Jaguar, Audi** has the **highest average selling price** as compare to Brands having High sales counts like **Maruti, Hyundai, Mahindra, Tata**.
